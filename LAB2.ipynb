```python
import numpy as np
import pandas as pd
from sklearn.datasets import *
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = data.drop(columns=data.columns[data.isna().sum() >= 3000])
```

```python
data.isna().sum()
```

```python
data = data.drop(["website", "reference", "wikipedia", "wikipedia_page_id", "wikipedia_appeared", "wikipedia_summary", "origin_community"], axis = 1)
```

```python
data.info()
```

```python
mode_value = data['central_package_repository_count'].mode()[0]
data['central_package_repository_count'] = data['central_package_repository_count'].fillna(mode_value)

mean_value = data['wikipedia_daily_page_views'].mean()
data['wikipedia_daily_page_views'] = data['wikipedia_daily_page_views'].fillna(mean_value)

mean_value = data['wikipedia_backlinks_count'].mean()
data['wikipedia_backlinks_count'] = data['wikipedia_backlinks_count'].fillna(mean_value)
```

```python
data.info()
```

```python
data = data.drop(["pldb_id", "title"], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['type_code'] = le.fit_transform(data['type'])
data = data.drop("type", axis = 1)
```

```python
numeric_cols = data.columns

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])
```
