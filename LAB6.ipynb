```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import (
    RandomForestRegressor,
    StackingRegressor
)

from sklearn.neural_network import MLPRegressor

import gmdh
```

```python
df = pd.read_csv("iphone_sales_dataset.csv")
```

```python
df.info()
```

```python
target_column = 'Price'

X = df.drop(columns=[target_column])
y = df[target_column]

if 'Sale_Date' in X.columns:
    X = X.drop(columns=['Sale_Date'])

categorical_columns = X.select_dtypes(include=['str']).columns

for col in categorical_columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

dt = DecisionTreeRegressor(max_depth=5, random_state=42)

dt.fit(X_train, y_train)

rf = RandomForestRegressor(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)
```

```python
estimators = [('dt', DecisionTreeRegressor(max_depth=5)),('rf',RandomForestRegressor(n_estimators=50))]

stacking_reg = StackingRegressor(estimators=estimators, final_estimator=LinearRegression())

stacking_reg.fit(X_train, y_train)
```

```python
mlp = MLPRegressor(
    hidden_layer_sizes=(100, 50),
    max_iter=1000,
    random_state=42
)

mlp.fit(X_train, y_train)
```

```python
print("\n" + "="*60)
print("РЕЗУЛЬТАТЫ МГУА")
print("="*60)

start_time = time.time()

combi_model = gmdh.MultilayerGMDH(ref_functions=gmdh.RefFunctionType.rfLinear)

combi_model.fit(X_train, y_train)
train_time_combi = time.time() - start_time

start_time = time.time()

mia_model = gmdh.MultilayerGMDH(
    ref_functions=(
        gmdh.RefFunctionType.rfLinearCov,
        gmdh.RefFunctionType.rfQuadratic,
        gmdh.RefFunctionType.rfCubic
    )
)
mia_model.fit(
    X_train,
    y_train
)
train_time_mia = time.time() - start_time
print('Модели МГУА обучены')
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression

estimators = [('dt', DecisionTreeRegressor(max_depth=5)), ('rf', RandomForestRegressor(n_estimators=50))]

stacking_reg = StackingRegressor(estimators=estimators, final_estimator=LinearRegression())

stacking_reg.fit(X_train, y_train)
```

```python
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)
```
